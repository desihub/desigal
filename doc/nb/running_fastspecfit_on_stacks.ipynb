# Running FastSpecFit on stacked spectra

To run fastspecfit you need to use the fastspecfit kernel. To set this up follow the instructions here: https://fastspecfit.readthedocs.io/en/latest/install.html#nersc-installation

Fastspecfit is used in the terminal, but here we will use the notebook to do this instead. There are two steps:
1. The actual fitting of the data (fastspecfit --> in this case the stackfit routine)
2. Producing QA figures to assess the quality of the fits (fastqa)

```python
import os
from astropy.table import Table
```

```python
def run_fastspecfit_on(filename):
    os.system('stackfit ' + filename + ' -o ' + 'stackfit_' + filename + ' --mp 1')
    return

def run_fastqa_on(filename):
    os.system('fastqa ' + 'stackfit_' + filename + ' --redrockfiles ' + filename + ' -o qa_plots/' + filename[:-5] + ' --stackfit')
    return
```

```python
filename = "example_stack.fits"
```

## Run FastSpecFit

```python
run_fastspecfit_on(filename)
```

### Reading the fitted parameters
The parameters can be read from the stackfit file which is formatted as explained here: https://fastspecfit.readthedocs.io/en/latest/fastspec.html

```python
Table.read('stackfit_'+filename, hdu=1)
```

## Run FastQA

```python
run_fastqa_on(filename)
```
